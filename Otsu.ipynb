```
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpig
from PIL import Image
```

```
aluminium = Image.open('aluminium.png')
aluminium
```

```
aluminium = cv2.imread('aluminium.png',0)

ret, imgf = cv2.threshold(aluminium, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.figure(figsize = (8,12))
plt.subplot(3,1,3), plt.imshow(imgf,cmap = 'gray')
plt.title('Otsu thresholding'), plt.xticks([]), plt.yticks([])
plt.imsave('aluminium_Otsu.png', imgf, cmap = 'gray')
plt.show()
print('Threshold value: ', ret)
```
